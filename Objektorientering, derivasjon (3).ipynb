Her skal vi se på fire ulike måter å tilnærme den deriverte på numerisk, og sammenligne med den analytiske derivasjonen. 

```python
from pylab import *


def f(x):
    return sin(2*pi*x)

def fder(x):
    return 2*pi*cos(2*pi*x)

class Diff:
    def __init__(self,f):
        self.f = f
        self.x=0
        self.h=0
    def forover(self, x, h):
        self.x = x
        self.h = h
        return (self.f(self.x+self.h)-self.f(self.x))/self.h
    
    def bakover(self,x, h):
        self.x = x
        self.h = h
        return(self.f(self.x)-self.f(self.x-self.h))/self.h
    
    def sentral(self, x, h):
        self.x = x
        self.h = h
        return (self.f(self.x+self.h)-self.f(self.x-self.h))/(2*self.h)

    def firepunkt(self, x, h):
        self.x = x
        self.h = h
        return (-self.f(self.x+2*self.h)+8*self.f(self.x+self.h)-8*self.f(self.x-self.h)+self.f(self.x-2*self.h))/(12*self.h)
    
    
    
derivert = Diff(f)

x = linspace(-1, 1, 100)
y = fder(x)


#Med h = 0.9
forover = derivert.forover(x, 0.9)
bakover = derivert.bakover(x, 0.9)
sentral = derivert.sentral(x, 0.9)
firepunkt = derivert.firepunkt(x, 0.9)


title("h = 0.9")
xlabel("x")
ylabel("fder(x)")
plot()
plot(x,y, label="Analytisk")
plot(x, forover, label="Forover")
plot(x, bakover, label="Bakover")
plot(x, sentral, label="Sentral")
plot(x, firepunkt, label="Firepunkt")
legend()
show()


#Med h = 0.6  
forover = derivert.forover(x, 0.6)
bakover = derivert.bakover(x, 0.6)
sentral = derivert.sentral(x, 0.6)
firepunkt = derivert.firepunkt(x, 0.6) 
        
title("h = 0.6")
xlabel("x")
ylabel("fder(x)")
plot()
plot(x,y, label="Analytisk")
plot(x, forover, label="Forover")
plot(x, bakover, label="Bakover")
plot(x, sentral, label="Sentral")
plot(x, firepunkt, label="Firepunkt")
legend()
show()


#Med h = 0.3
forover = derivert.forover(x, 0.3)
bakover = derivert.bakover(x, 0.3)
sentral = derivert.sentral(x, 0.3)
firepunkt = derivert.firepunkt(x, 0.3) 
        
title("h = 0.3")
xlabel("x")
ylabel("fder(x)")
plot()
plot(x,y, label="Analytisk")
plot(x, forover, label="Forover")
plot(x, bakover, label="Bakover")
plot(x, sentral, label="Sentral")
plot(x, firepunkt, label="Firepunkt")
legend()
show()


#Med h = 0.1
forover = derivert.forover(x, 0.1)
bakover = derivert.bakover(x, 0.1)
sentral = derivert.sentral(x, 0.1)
firepunkt = derivert.firepunkt(x, 0.1) 
        
title("h = 0.1")
xlabel("x")
ylabel("fder(x)")
plot()
plot(x,y, label="Analytisk")
plot(x, forover, label="Forover")
plot(x, bakover, label="Bakover")
plot(x, sentral, label="Sentral")
plot(x, firepunkt, label="Firepunkt")
legend()
show()
```

Som vi ser på plottene, vil den numeriske tilnærmingen av den deriverte avhenge mye av hvilken h-verdi man velger. Vi ser at når man velger h = 0.9 vil punktene være for langt unna hverandre, så den numerisk deriverte blir veldig unøyaktig. Vi ser at jo mindre h-verdi vi velger, jo nærmere den analytiske verdien kommer vi. Derfor er det viktig å velge en liten h-verdi når man deriverer numerisk, ellers vil derivasjonen blir veldig unøyaktig, slik som vi ser i det første plottet, der h = 0.9.
